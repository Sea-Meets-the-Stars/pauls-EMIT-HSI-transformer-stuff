```python
import subprocess
import os 
import glob
s3_bucket = "s3://EMIT_Hyperspectral_Proj_Killam_justcubes/"
destination_folder = "/root/emit_data"
wildcard_pattern = "20230107*.tar"
```

```python
os.makedirs(destination_folder, exist_ok=True)

#pull the data from s3
s3_cmd = f"aws s3 cp {s3_bucket} {destination_folder} --recursive --exclude '*.tar' --include '{wildcard_pattern}'"
subprocess.run(s3_cmd, shell=True)
```

```python
#untar the data
tar_file_list = glob.glob(os.path.join(destination_folder, "*.tar"))

for tar_file in tar_file_list:
    untar_cmd = f"tar -xf {tar_file} -C {destination_folder}"
    subprocess.run(untar_cmd, shell=True)
```

```python
#remove the tar files
for tar_file in tar_file_list:
    os.remove(tar_file)
```
