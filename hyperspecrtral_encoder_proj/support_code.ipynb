```python
import numpy as np
import glob
```

```python
def inspect_emit_npy(path, band_dim=285):
    assert "RAD" in path, f"Refusing non-RAD file: {path}"
    x = np.load(path, mmap_mode="r")  # avoids loading whole file into RAM
    xf = x[np.isfinite(x)]
    print(f"path:  {path}")
    print(f"shape: {x.shape}   dtype: {x.dtype}")
    if x.ndim == 3:
        dims = list(x.shape)
        if band_dim in dims:
            b = dims.index(band_dim)
            print(f"band_dim index: {b}  (expected {band_dim} bands)")
    print(f"finite: {xf.size}/{x.size}  nan: {np.isnan(x).sum()}  inf: {np.isinf(x).sum()}")
    if xf.size:
        p = np.percentile(xf, [0, 0.1, 1, 50, 99, 99.9, 100])
        print(f"min/mean/std/max: {xf.min():.6g}  {xf.mean():.6g}  {xf.std():.6g}  {xf.max():.6g}")
        print(f"p0/p0.1/p1/p50/p99/p99.9/p100: {p}")
        
inspect_emit_npy("/dataset/CSE290D_F25/data/emit_dataset/chips/ch4_plume/PLUME_EMIT_L1B_RAD_001_20230107T143818_2300709_004_y1280_x256_l1b.npy")
```
